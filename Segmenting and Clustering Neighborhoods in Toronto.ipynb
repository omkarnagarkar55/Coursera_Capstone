```python
import pandas as pd
import numpy as np
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
```

# Ignore SSL certificate errors

```python
url=r"https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
url
```

```python
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
all=soup.find_all("table",{'class':'wikitable sortable'})
all  
```

```python
columns=[]
for i in all[0].find_all("th"):
    columns.append(i.text.rstrip('\n'))
columns  
```

```python
l=[]
for i in all[0].find_all('tr'):
    l.append(i.text.replace('\n\n\n',',').replace('\n\n',',').split(','))
l    
```

```python
l=l[1:]
neighborhoods = pd.DataFrame(columns=columns)
neighborhoods
```

```python
for i in l:  
    pc=i[0].replace('\n','')
    b=i[1]
    n=i[2].replace('\n','').replace('/',',')
    neighborhoods = neighborhoods.append({'Postal code':pc,
                                          'Borough':b,
                                          'Neighborhood':n},
                                           ignore_index=True)
neighborhoods                                    
```

```python
hood=neighborhoods[neighborhoods['Borough']!='Not assigned'].reset_index()
hood
```

```python
geo=pd.read_csv('http://cocl.us/Geospatial_data',index_col=None)
geo.columns=['Postal code','Latitude','Longitude']
geo
```

```python
hood1 = pd.merge(hood,geo,on='Postal code').drop(['index'],axis=1)
hood1
```
